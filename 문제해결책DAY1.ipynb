###수치형 데이터 시각화

```python
import seaborn as sns

titanic = sns.load_dataset('titanic') # 타이타닉 데이터 불러오기
```

```python
titanic.head() # 데이터가 어떻게 구성되어있는지 살펴보는것

# 수치형 데이터와 범주형 데이터가 공존하고있는것을 확인

#histplot() : 히스토그램
#kdeplot() : 커널밀도추정 함수 그래프
#displot() : 분포도
#rugplot() : 러그플롯
```

```python
sns.histplot(data=titanic, x='age');
```

```python
sns.histplot(data=titanic, x='age', bins=10); 

# 히스토그램은 기본적인 수치형 데이터 하나에 대한 빈도를 나타냄
```

```python
sns.histplot(data=titanic, x='age', hue='alive');

#생존여부에 따른 연령분포를 그려주는 코드

# 생존자와 사망자수 그래프를 포개서 나타냄 회색구간이 겹친부분

```

```python
sns.histplot(data=titanic, x='age', hue='alive', multiple='stack')

# 생존자와 사망자를 누적으로 표현함 누적은 multiple='stack'
```

```python
sns.kdeplot(data=titanic, x='age');

# 커널밀도추정 함수 그래프는 탐색적 데이터를 분석시 많이 쓰임
```

```python
sns.kdeplot(data=titanic, x='age', hue='alive', multiple='stack');

#히스토그램을 매끄럽게 연결한 느낌이 드는 그래프

```

```python
sns.displot(data=titanic, x='age');

# 히스토그램 과 비교하면 크가만 다를뿐
```

```python
sns.displot(data=titanic, x='age', kind='kde');
```

```python
sns.displot(data=titanic, x='age', kde=True);

# 히스토그램과 커널밀도 추정함수 그래프를 동시에 그릴수도 있다
```

```python
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age');

#러그플롯은 주변분포를 나타내는 그래프

#다른 분포도 그래프와 같이 쓰인다

#러그플롯은 (age 피처)가 어떻게 분포돼 있는지를
#X축 위에 작은 선분으로 표시됨
```

```python
sns.barplot(x='class', y='fare', data=titanic);

#막대그래프는 범주형 데이터 값에 따라 

#수치형 데이터 값이 어떻게 달라지는지 파악할때 사용

#수치형은 막대높이로, 신뢰구간은 오차막대로 표현

#샘플링한 데이터 평균을 구해줌

#막대높이는 등급별 평균운임, 막대상단 검은색 세로줄은 오차막대

# 등급이 높을수록 평균운임이 비싸고 신뢰구간이 넓어짐
```

```python
sns.pointplot(x='class', y='fare', data=titanic);

# 막대그래프와 모양만 다를뿐 동일한 정보를 제공함

# 한화면에 여러개의 그래프를 그릴때 유용함

```

```python
sns.boxplot(x='class', y='age', data=titanic);

#박스플롯은 막대그래프나 포인트플롯보다 많은 정보를 제공가능

#제1사분위 수(Q1) : 전체 데이터 중 하위 25%에 해당하는 값
#제2사분위 수(Q2) : 50% 해당하는 값(중앙값)값
#제3사분위 수(Q3) : 상위 25%에 해당하는 값
#사분위 범위 수(IQR) : Q3 - Q1
#최댓값 : Q3 +(1.5*IQR)
#최소값 : Q1 -(1.5*IQR)
#이상치 : 최댓값보다 큰 값과 최솟값보다 작은 값
```

```python
sns.violinplot(x='class', y='age', data=titanic);

#바이올린플롯은 박스플롯과 커널밀도추정함수 그래프를 합쳐놓은 그래프

#박스플롯이 제공하는 정보를 모두 포함 
```

```python
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True)

#성별에 따른 등급별 나이분포를 나타냄
```

```python
sns.countplot(x='class', data=titanic);

#범주형 데이터의 개수를 파악할때 쓰는그래프

# 범주형 피처나 타깃값의 분포가 어떤지 파악용도로 사용
```

```python
sns.countplot(y='class', data=titanic);

# 파라미터를 x를 y로 바꾸면 그래프 방향이 바뀐다
```

```python
sns.barplot(x='class', y='fare', data=titanic); # 막대그래프
sns.countplot(y='class', data=titanic);         # 카운트플롯
```
