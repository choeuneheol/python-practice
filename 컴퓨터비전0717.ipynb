```python
#이미지 바차트로 띄우기


img = cv2.imread('images/puppy.bmp',0)
bgr = cv2.imread('images/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()
    
img_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))

plt.subplot(131), plt.imshow(bgr), plt.axis('off')
plt.subplot(132), plt.imshow(img_rgb), plt.axis('off')
plt.subplot(133), plt.imshow(img, cmap='gray'), plt.axis('off')
    

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 알파채널 만들어서 강아지에 안경씌워보기

src = cv2.imread('fig/fig/puppy.bmp')
sunglass = cv2.imread('fig/fig/imgbin_sunglasses_1.png', cv2.IMREAD_UNCHANGED)
sunglass = cv2.resize(sunglass,(300,150))

mask = sunglass[:,:,-1]
glass = sunglass[:,:,0:3]

h,w = mask.shape[:2]

src_crop = src[120:120+h, 220:220+w]

cv2.copyTo(glass, mask, src_crop)

cv2.imshow('src',src)
cv2.imshow('mask',mask)
cv2.imshow('glass',glass)
cv2.imshow('crop',src_crop)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#선그리기

img = np.full((600,800,3), (255,255,255), dtype=np.uint8)

cv2.imshow('image',img)

cv2.line(img,(50,50),(300,50),(0,0,0),4)
cv2.line(img,(170,50),(200,250),(0,0,0),4)
cv2.line(img,(170,50),(130,250),(0,0,0),4)

cv2.rectangle(img,(50,400),(400,500),(0,0,0),20)

cv2.rectangle(img,(60,450,300,150),(255,255,255),-1)

cv2.circle(img,(500,300),100,(0,0,255),10,cv2.LINE_AA)

cv2.ellipse(img,(600,300),(50,100),30,0,100,(171,0,255),10,cv2.LINE_AA)

cv2.putText(img, "WOW!",(600,500),cv2.FONT_HERSHEY_COMPLEX,2,(133,170,200),5,cv2.LINE_AA)

cv2.imshow('img',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
